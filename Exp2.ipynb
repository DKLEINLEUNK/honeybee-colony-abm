```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from continuous_model.Bee import BeeSwarm
from continuous_model.Hive import Hive
from continuous_model.Model import ForagerModel
from continuous_model.Resource import Resource
from continuous_model.config import *
from server import agent_potrayal, model_params

model_config = ModelConfig(size=2000)
beeswarm_config = BeeSwarmConfig()
hive_config = HiveConfig()
resource_config = ResourceConfig()
model = ForagerModel(1, 10, model_config, hive_config, beeswarm_config, resource_config)
# model.size = 2000
```

```python
def plot_space(model):
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_xlim(0, model.size)
    ax.set_ylim(0, model.size)
    for agent in model.schedule.all_agents[BeeSwarm].values():
        ax.scatter(agent.pos[0], agent.pos[1], c="yellow", s=100)
    for agent in model.schedule.all_agents[Resource].values():
        ax.scatter(agent.pos[0], agent.pos[1], c="green", s=100)
    for agent in model.schedule.all_agents[Hive].values():
        ax.scatter(agent.pos[0], agent.pos[1], c="black", s=100)
    plt.show()
plot_space(model)
```

```python
%%time
# Run 500 steps
for t in range(200):
    model.step()
```
